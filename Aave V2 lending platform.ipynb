```python
import json

# Load the JSON file
with open("user-wallet-transactions.json", "r") as f:
    data = json.load(f)

# Check the total number of transactions
print(f"Total transactions loaded: {len(data)}")

# Show one sample record
print("Sample transaction record:")
print(data[0])
```

```python
import pandas as pd

# Convert the JSON data to a DataFrame
df = pd.DataFrame(data)

# View the top 5 rows
df.head()
```

```python
def extract_amount_usd(action_data):
    try:
        if isinstance(action_data, dict):
            raw_amount = float(action_data.get('amount', '0'))
            price = float(action_data.get('assetPriceUSD', '1'))  # fallback to 1 if missing
            amount_usd = raw_amount * price
            # Most tokens have 18 decimals, adjust if needed
            return amount_usd / (10 ** 18)
    except:
        return 0.0
```

```python
df['amount_usd'] = df['actionData'].apply(extract_amount_usd)

# Check result
df[['action', 'amount_usd']].head()
```

```python
wallet_df = df.groupby(['userWallet', 'action'])['amount_usd'].sum().unstack(fill_value=0).reset_index()

# Add count of liquidationCalls
liquidation_counts = df[df['action'] == 'liquidationCall'].groupby('userWallet').size().reset_index(name='liquidationCount')

# Merge both
wallet_df = wallet_df.merge(liquidation_counts, on='userWallet', how='left')
wallet_df['liquidationCount'] = wallet_df['liquidationCount'].fillna(0)

wallet_df.head()
```

```python
for col in ['deposit', 'borrow', 'repay', 'redeemUnderlying']:
    if col not in wallet_df.columns:
        wallet_df[col] = 0
```

```python
def calculate_score(row):
    base_score = 600
    
    borrow = row['borrow']
    repay = row['repay']
    deposit = row['deposit']
    liquidations = row['liquidationCount']
    
    repay_ratio = repay / borrow if borrow > 0 else 1
    deposit_bonus = min(deposit / 100, 100)  # up to 100 bonus points
    liquidation_penalty = liquidations * 100

    score = base_score + (repay_ratio * 200) + deposit_bonus - liquidation_penalty
    return min(max(int(score), 0), 1000)

wallet_df['score'] = wallet_df.apply(calculate_score, axis=1)

wallet_df[['userWallet', 'score']].head()
```

```python
wallet_df[['userWallet', 'score']].to_csv("wallet_scores.csv", index=False)
```

```python
import pandas as pd

# Flatten actionData
action_data = pd.json_normalize(df['actionData'])

# Merge with main DataFrame
df_flat = pd.concat([df.drop(columns=['actionData']), action_data], axis=1)

# Convert relevant fields to numeric
df_flat['amount'] = pd.to_numeric(df_flat['amount'], errors='coerce')
df_flat['assetPriceUSD'] = pd.to_numeric(df_flat['assetPriceUSD'], errors='coerce')

# Compute amount in USD
df_flat['amount_usd'] = df_flat['amount'] * df_flat['assetPriceUSD']

# Convert timestamp
df_flat['timestamp'] = pd.to_datetime(df_flat['timestamp'])

# Filter valid rows
df_flat = df_flat.dropna(subset=['amount_usd', 'type'])

# Preview
df_flat.head()
```

```python
selected_columns = [
    'userWallet', 'type', 'amount_usd', 'timestamp',
    'borrowRate', 'repayerId', 'liquidatorId',
    'collateralAmount', 'principalAmount', 'assetSymbol'
]

df_selected = df_flat[selected_columns].copy()
```

```python
# Total USD amounts per action type per user
pivot_amounts = df_selected.pivot_table(
    index='userWallet',
    columns='type',
    values='amount_usd',
    aggfunc='sum',
    fill_value=0
)

# Total number of each transaction type per user
pivot_counts = df_selected.pivot_table(
    index='userWallet',
    columns='type',
    values='amount_usd',
    aggfunc='count',
    fill_value=0
)

# Rename columns for clarity
pivot_amounts.columns = [f'total_{col.lower()}_usd' for col in pivot_amounts.columns]
pivot_counts.columns = [f'count_{col.lower()}' for col in pivot_counts.columns]

# Merge them together into one wallet-level feature set
wallet_features = pd.concat([pivot_amounts, pivot_counts], axis=1)

# Fill missing values with 0 just in case
wallet_features.fillna(0, inplace=True)

# Reset index so 'userWallet' becomes a normal column
wallet_features.reset_index(inplace=True)

wallet_features.head()
```

```python
import pandas as pd

# Ensure numeric conversion
df_selected['amount_usd'] = pd.to_numeric(df_selected['amount_usd'], errors='coerce').fillna(0)
df_selected['borrowRate'] = pd.to_numeric(df_selected['borrowRate'], errors='coerce')

# Add action flags
df_selected['is_deposit'] = df_selected['type'] == 'Deposit'
df_selected['is_borrow'] = df_selected['type'] == 'Borrow'
df_selected['is_repay'] = df_selected['type'] == 'Repay'
df_selected['is_withdraw'] = df_selected['type'] == 'RedeemUnderlying'
df_selected['is_liquidated'] = df_selected['type'] == 'LiquidationCall'

# Group by userWallet
wallet_features = df_selected.groupby('userWallet').agg(
    total_transactions=('type', 'count'),
    total_deposit_usd=('amount_usd', lambda x: x[df_selected.loc[x.index, 'is_deposit']].sum()),
    total_borrow_usd=('amount_usd', lambda x: x[df_selected.loc[x.index, 'is_borrow']].sum()),
    total_repay_usd=('amount_usd', lambda x: x[df_selected.loc[x.index, 'is_repay']].sum()),
    total_withdraw_usd=('amount_usd', lambda x: x[df_selected.loc[x.index, 'is_withdraw']].sum()),
    times_liquidated=('is_liquidated', 'sum'),
    average_borrow_rate=('borrowRate', 'mean'),
).reset_index()

# Fill NaNs (optional safety step)
wallet_features.fillna(0, inplace=True)

# Derived behavior ratios
wallet_features['borrow_deposit_ratio'] = wallet_features['total_borrow_usd'] / (wallet_features['total_deposit_usd'] + 1)
wallet_features['repay_borrow_ratio'] = wallet_features['total_repay_usd'] / (wallet_features['total_borrow_usd'] + 1)

wallet_features.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Select relevant features for clustering
features = wallet_df[['deposit', 'borrow', 'repay', 'redeemUnderlying', 'liquidationCount']]

# Fill missing values if any
features = features.fillna(0)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)

# Apply KMeans clustering (k=3 can be adjusted later)
kmeans = KMeans(n_clusters=3, random_state=42)
wallet_df['cluster'] = kmeans.fit_predict(X_scaled)

# View distribution
print(wallet_df['cluster'].value_counts())

# Optional: Visualize clusters
sns.pairplot(wallet_df, hue='cluster', vars=['deposit', 'borrow', 'repay', 'liquidationCount'])
plt.suptitle("User Clustering Based on Financial Behavior", y=1.02)
plt.show()
```

```python
# Normalize ratios between 0 and 1
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
wallet_features[['borrow_deposit_ratio_scaled', 'repay_borrow_ratio_scaled']] = scaler.fit_transform(
    wallet_features[['borrow_deposit_ratio', 'repay_borrow_ratio']]
)

# Optional: cluster-based weight
cluster_weights = {
    0: 1.0,  # safest cluster
    1: 0.7,
    2: 0.4   # risky users
}

wallet_features['cluster_weight'] = wallet_features['cluster'].map(cluster_weights)

# Final score out of 1000
wallet_features['score'] = (
    (wallet_features['repay_borrow_ratio_scaled'] * 0.6 +
     (1 - wallet_features['borrow_deposit_ratio_scaled']) * 0.4)
    * wallet_features['cluster_weight']
    * 1000
).astype(int)
```
